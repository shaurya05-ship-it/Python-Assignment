```python
# PROBLEM 1
# PART 1

def GCD(m,n):
    while n!=0:
        m,n=n,m % n
    return m
GCD(2,3)
```

```python
# PART 2
def GCD(m,n):
    m,n= abs(m),abs(n)#If one or both inputs are negative, the function still works, but the result may be negative depending on the inputs.
    m,n= abs(m),abs(n)
    while n!=0:
        m,n=n,m % n
    return m

GCD(-2,-3)
```

```python
# PROBLEM 2
#PART 1
def euler_limit(n):
    return float((1+1/n)**n)

euler_limit(268)
```

```python
# PART 2
#when n is extremely large, the approximation does not improve.
#because computers use floating point numbers and which cannot represent extremely small numbers precisely.
#When the number 1/x gets too small, the computer will interpret this as:
#1+1/x ~ 1
#so the result becomes inaccurate
```

```python
# PART 3
def factorial(v):
    if v==0 or v==1:
        return 1
    return v*factorial(v-1)

factorial(5)    
```

```python
def euler_sum(n):
    if n==0:
        return 0.0

    total=0.0
    for v in range(n):
        total+=1/factorial(v)

    return float(total)

euler_sum(3)
```

```python
# PART 4
import math

def euler_approx(epsilon):
    total=0.0
    k=0
    
    while True:
        total+=1/factorial(k)
        if abs(total-math.exp(1))<epsilon:
            return total
            k+=1
            
euler_approx(5)
```

```python
# PART 5
import math

def print_euler_sum_error(n):
    true_e = math.exp(1)
    for k in range(1, n+1):
        error = euler_sum(k) - true_e
        print(error)
print_euler_sum_error(7)
```

```python
def print_euler_lim_error(n):
    true_e = math.exp(1)
    for k in range(1, n+1):
        error = euler_limit(k) - true_e
        print(error)
print_euler_lim_error(5)
```

```python
# PROBLEM 3
#PART 1
def is_power_of_2(n):
    if n==1:
        return True
    if n%2!=0:
        return False
    return is_power_of_2(n//2)
is_power_of_2(4)
```

```python
#PART 2
def is_power(b, n):
    if n == 1:
        return True
    if n % b != 0:
        return False
    return is_power(b, n // b)
is_power(12,23)
```
